# Monty Hall Lab 1

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
door_options    = np.array([1,2,3])
door_selection  = np.array([1,2,3])
count_stay      = 0
count_switch    = 0
stay            = []
switch          = []
```

```python
for it in range(1000):
    random.shuffle(door_options)
    random.shuffle(door_selection)
    winner_door     = door_options[0]
    sel_door        = door_selection[0]
    morty_door      = door_selection[1] if winner_door != door_selection[1] else door_selection[2]

    count_stay   = (count_stay+1) if winner_door == sel_door else count_stay
    if(morty_door == door_selection[1]):
        count_switch = (count_switch+1) if winner_door == door_selection[2] else count_switch
    else:
        count_switch = (count_switch+1) if winner_door == door_selection[1] else count_switch
    
    stay.append(count_stay)
    switch.append(count_switch)
```

```python
plt.plot(stay,label="Stay",color='red')
plt.plot(switch,label="Switch",color='green')
#plt.plot(random_choice,label="Switch",color='green')
plt.legend()
plt.title("Monty Hall")
plt.show()
```
